```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# angle
theta = np.pi / 6   # 30 degrees

# rotation matrix Q
Q = np.array([[np.cos(theta), -np.sin(theta)],
              [np.sin(theta),  np.cos(theta)]])

# basis vectors
e1 = np.array([1, 0])
e2 = np.array([0, 1])

# rotated vectors
Qe1 = Q @ e1
Qe2 = Q @ e2

# plot setup
fig, ax = plt.subplots()
ax.set_aspect('equal')
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
ax.axhline(0, color='gray', linewidth=0.5)
ax.axvline(0, color='gray', linewidth=0.5)

# function to draw vector and annotate
def draw_vector(v, color, label, text):
    ax.plot([0, v[0]], [0, v[1]], '-', color=color, linewidth=2, label=label)
    ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1,
              color=color, alpha=0.8)
    ax.text(v[0]*1.1, v[1]*1.1, text, fontsize=12, ha='center', va='center')

# original basis (darker colors)
draw_vector(e1, 'blue', '$e_1$', '(1,0)')
draw_vector(e2, 'green', '$e_2$', '(0,1)')

# rotated basis (lighter colors)
draw_vector(Qe1, 'lightskyblue', '$Qe_1$', '$(\cos\\theta, \sin\\theta)$')
draw_vector(Qe2, 'lightgreen', '$Qe_2$', '$(-\sin\\theta, \cos\\theta)$')

# theta arc for e1 → Qe1
arc1 = patches.Arc((0,0), 0.6, 0.6, angle=0,
                   theta1=0, theta2=np.degrees(theta),
                   color="black")
ax.add_patch(arc1)
ax.text(0.7*np.cos(theta/2), 0.7*np.sin(theta/2), r"$\theta$", fontsize=14)

# theta arc for e2 → Qe2
arc2 = patches.Arc((0,0), 0.6, 0.6, angle=0,
                   theta1=90, theta2=90+np.degrees(theta),
                   color="black")
ax.add_patch(arc2)
ax.text(0.7*np.cos(np.pi/2 + theta/2), 0.7*np.sin(np.pi/2 + theta/2), r"$\theta$", fontsize=14)

# legend and title
ax.legend()
plt.title("Rotation of Basis Vectors by θ")
plt.show()
```
